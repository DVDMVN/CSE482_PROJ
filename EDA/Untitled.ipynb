```python
canon_grouped = (
    canon.groupby('title')
    .agg(
        positive_count=('sentiment_total', lambda x: (x > 0).sum()),
        negative_count=('sentiment_total', lambda x: (x < 0).sum()),
        total_count=('sentiment_total', 'count'),
        mean=('sentiment_total', 'mean'),
        Overall_sentiment=('sentiment_total', 'sum'),
    )
    .reset_index()
)

canon_grouped
```

```python
import nltk
import spacy
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from spacy.lang.en import English

# Load the spaCy model
nlp = spacy.load('en_core_web_sm')

# Download the NLTK stopwords if you haven't already
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')



# Sample dataset (replace with your dataset)
data = canon['sentence']

# Initialize the stemmer
stemmer = PorterStemmer()

# Preprocessing function using NLTK and spaCy
def preprocess_text(text):
    # Lowercase the text
    text = text.lower()
    
    # Remove punctuation and special characters using regular expressions
    text = re.sub(r'[^\w\s]', '', text)
    
    # Tokenize the text using NLTK
    tokens = word_tokenize(text)
    
    # Remove stopwords using NLTK
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    
    # Lemmatization using spaCy
    doc = nlp(" ".join(tokens))  # Rejoin tokens before processing with spaCy
    lemmatized_tokens = [token.lemma_ for token in doc]
    
    # Optionally, apply stemming using NLTK
    # Uncomment below line if you want to use stemming instead of lemmatization
    # lemmatized_tokens = [stemmer.stem(word) for word in lemmatized_tokens]
    
    # Return the preprocessed text as a space-separated string of words
    return " ".join(lemmatized_tokens)

# Apply the preprocessing to the dataset
preprocessed_data = [preprocess_text(text) for text in data]

# Display the results
for original, preprocessed in zip(data, preprocessed_data):
    print(f"Original: {original}")
    print(f"Preprocessed: {preprocessed}")
    print("-" * 50)
```
